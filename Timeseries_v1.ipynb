```python
#import numpy as np # we will use this later, so import it now
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook, output_file, show
#from bokeh.io import output_file, show
output_notebook()
import pandas as pd
import numpy as np

df_2018 = pd.read_csv('2018.csv')
#print(df_2018.head())
my_list = df_2018.columns.values.tolist()
print(my_list)
#print(df_2018['Health_Exp'].isnull().values.any()) #there are null values to consider
#df_2018.plot() #shows which of the variables are considered numbers visually
#print(df_2018.dtypes)

# create percentage variables, hint use np.divide()   
df_2018['per_police_exp'] = (np.divide(df_2018["Police_Exp"], df_2018["Tot_Exp"])*100)
df_2018['per_health_exp'] = (np.divide(df_2018["Health_Exp"], df_2018["Tot_Exp"])*100)
df_2018['per_housing_exp'] = (np.divide(df_2018["Housing_Exp"], df_2018["Tot_Exp"])*100)
df_2018['pp_police_exp'] = (np.divide(df_2018["Police_Exp"], df_2018["Population"]))
df_2018['pp_health_exp'] = (np.divide(df_2018["Health_Exp"], df_2018["Population"]))
df_2018['pp_housing_exp'] = (np.divide(df_2018["Housing_Exp"], df_2018["Population"]))

df_2018['pp_tot_exp'] = (np.divide(df_2018["Tot_Exp"], df_2018["Population"]))
#print(df_2018.head())

###next steps
##join datasets
##compare police_exp_pp (x) versus per_white
##

##attempting via directly calling variables
p = figure(plot_width=800, plot_height=400, title="My Plot")
#p.circle(df_2018['Tot_Rev'], df_2018['Tot_Exp'])
p.circle(df_2018['pp_police_exp'], df_2018['pp_tot_exp'])
#p.circle(df_2018['per_police_exp'], df_2018['Entity_Name']) #doesn't show, maybe because an object and not a string
output_file("plot.html")
show(p)

#output_file(f"C:/Users/mwrable/Desktop/Police_Project/Bokeh_test.html")

##attempting via ColumnDataSource method
#source = ColumnDataSource(data=df_2000)
#print(source[' CITY2 '])
#p = figure()
#p.circle(x=' TOTAL_REV ', y=' TOTAL_EXP ', source=source)
#show(p)

```
